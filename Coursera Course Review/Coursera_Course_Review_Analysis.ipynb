Course Paling Populer

Terlihat bahwa course machine learning merupakan course paling populer di Coursera

Kita akan melihat rating-rating yang didapatkan dari 3 course paling populer yaitu machine learning, learning how to learn, dan python

```
plt.rcParams['figure.figsize'] = (18, 10)

mlrating = reviewbycourse['Label'][reviewbycourse['CourseId'] == 'machine-learning']
plt.subplot(1, 3, 1)
labels = ['5', '4', '3', '2', '1']
colors = plt.cm.autumn_r(np.linspace(0, 1, 5))
mlrating.value_counts().plot.pie(colors=colors, labels = labels, explode=[0, 0, 0, 0.3, 0.5], autopct='%.2f%%', shadow=True, labeldistance=None)
plt.axis('off')
plt.title('Rating Untuk Kelas \n\n Machine Learning', fontsize=15)
plt.legend()

lhtlrating = reviewbycourse['Label'][reviewbycourse['CourseId'] == 'learning-how-to-learn']
plt.subplot(1, 3, 2)
colors = plt.cm.winter_r(np.linspace(0, 1, 5))
lhtlrating.value_counts().plot.pie(colors=colors, labels=labels, explode=[0, 0, 0, 0.3, 0.5], autopct='%.2f%%', shadow=True, labeldistance=None)
plt.axis('off')
plt.title('Rating Untuk Kelas \n\n Learning How To Learn', fontsize=15)
plt.legend()

pyrating = reviewbycourse['Label'][reviewbycourse['CourseId'] == 'python']
plt.subplot(1, 3, 3)
colors = plt.cm.Wistia(np.linspace(0, 1, 5))
pyrating.value_counts().plot.pie(colors=colors, labels=labels, explode=[0, 0, 0, 0.3, 0.5], autopct='%.2f%%', shadow=True, labeldistance=None)
plt.axis('off')
plt.title('Rating Untuk Kelas Python', fontsize=15)
plt.legend()

plt.show()
```

Terlihat bahwa 3 kelas terpopuler yakni Machine Learning, Learning How To Learn, dan Python mendapatkan mayoritas rating 5
