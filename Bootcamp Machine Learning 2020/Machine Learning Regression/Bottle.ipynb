```
df.describe()
```

```
from sklearn.preprocessing import StandardScaler, PowerTransformer, QuantileTransformer, MinMaxScaler, MaxAbsScaler, RobustScaler
```

```
stdr = StandardScaler()
powbox = PowerTransformer(method='yeo-johnson', standardize=True)
powbox1 = PowerTransformer(method='box-cox', standardize=True)
quan = QuantileTransformer()
minmax = MinMaxScaler()
maxabs = MaxAbsScaler()
rbst = RobustScaler()
```

```
X = df.drop(['T_degC'], axis=1)
y = df['T_degC']
```

```
X_stdr = stdr.fit_transform(X.values)
y_stdr = stdr.fit_transform(df['T_degC'].values.reshape(-1,1)).flatten()

X_powbox = powbox.fit_transform(X.values)
y_powbox = powbox.fit_transform(df['T_degC'].values.reshape(-1,1)).flatten()

#X_powbox1 = powbox1.fit_transform(X.values)
#y_powbox1 = powbox1.fit_transform(df['T_degC'].values.reshape(-1, 1)).flatten()

X_quan = quan.fit_transform(X.values)
y_quan = quan.fit_transform(df['T_degC'].values.reshape(-1,1)).flatten()

X_minmax = minmax.fit_transform(X.values)
y_minmax = minmax.fit_transform(df['T_degC'].values.reshape(-1,1)).flatten()

X_maxabs = maxabs.fit_transform(X.values)
y_maxabs = maxabs.fit_transform(df['T_degC'].values.reshape(-1,1)).flatten()

X_rbst = rbst.fit_transform(X.values)
y_rbst = rbst.fit_transform(df['T_degC'].values.reshape(-1,1)).flatten()
```

Power Transformers dengan metode Box-Cox tidak dapat digunakan karena pada dataset ditemukan data dengan nilai <= 0

### **Linear Regression**

**Standard Scaler**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_stdr, y_stdr, test_size=0.25, random_state=10)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
```

```
linreg = LinearRegression()
linreg.fit(X_train, y_train)
prediksi_linreg = linreg.predict(X_test)
```

```
mse = mean_squared_error(y_test, prediksi_linreg)
mae = mean_absolute_error(y_test, prediksi_linreg)
r2 = r2_score(y_test, prediksi_linreg)
print('MSE (Mean Squared Error) Using Standard Scaler is  :', mse)
print('MAE (Mean Absolute Error) Using Standard Scaler is :', mae)
print('r^2 score Using Standard Scaler is                 :', r2)
print('RMSE (Root Mean Squared Error) is                  :', np.sqrt(mean_squared_error(y_test, prediksi_linreg)))
```

**K-Nearest Neighbors Regressor**

```
from sklearn.neighbors import KNeighborsRegressor
```

```
neig = np.arange(1, 10, 2)
scorer2 = []

for i, k in enumerate(neig):
  knn = KNeighborsRegressor(n_neighbors=k)
  knn.fit(X_train_maxabs, y_train_maxabs)
  prediksi_knn_test = knn.predict(X_test_maxabs)
  scorer2.append(r2_score(y_test_maxabs, prediksi_knn_test))

plt.figure(figsize=[10,8])
plt.plot(neig, scorer2, label = 'Testing R2')
plt.legend()
plt.title('-value VS R2')
plt.xlabel('Number of Neighbors')
plt.ylabel('R2')
plt.xticks(neig)
plt.grid()
plt.show()
print("\nNilai R2 terbaik untuk Model KNN adalah {}".format(round(np.max(scorer2), 2)))
```

Dengan model Random Forest, Support Vector Regressor, dan XGBoost Regressor diperoleh nilai rmse yang sangat mendekati nol dan nilai r2 yang sangat tinggi (yakni diatas 0.95), dengan nilai r2 tertinggi diperoleh dengan model Linear Regression, XGBoost Regressor, dan Random Forest Regressor. 

Untuk model KNN belum sempat dirun jadi tidak diketahui nilai r2 dan rmsenya
